```python
# Print Odd Numbers in the list
numbers = (20, 10, 16, 19, 25, 1, 276, 188)

for number in numbers:
  if number % 2 == 1:
    print(number, "is odd")
```

```python
# Print Even numbers in the list
numbers = (20, 10, 16, 19, 25, 1, 276, 188)

for number in numbers:
  if number % 2 == 0:
    print(number, "is even")
```
