```python
# Body Mass Index

# Caputre the user input and cast to float
bmi = float(input("Enter the BMI Index: "))

if bmi < 18.5:
  print("Underweight")
elif bmi < 24.9:
  print("Normal")
elif bmi < 29.9:
  print("Overweight")
elif bmi < 34.9:
  print("Obesity Mild (Class 1)")
elif bmi < 39.9:
  print("Obesity Moderate (Class 2)")
else:
  print("Obesity Morbid (Class 3)")
```
